### 3. Data Cleaning
###### Removed duplicates to ensure accurate analysis. No missing values were present.

```python
# Here we fill missing numeric columns with their median
df_cleaned = df.fillna(df.median(numeric_only=True))
df
```

```python
# Drop duplicate entries if any
df_cleaned = df.drop_duplicates()
df_cleaned
```

```python
# Confirm cleaning steps
print("\nMissing values after cleaning:\n", df_cleaned.isnull().sum())
```

```python
print("\nDataset shape after cleaning:", df_cleaned.shape)
```

```python
# Save cleaned data if needed
df.to_csv('student-mat-cleaned.csv', index=False)
```

### 4. Data Analysis Questions

#### Q1: Average final grade (G3)

```python
avg_g3 = df_cleaned['G3'].mean()
avg_g3
```

#### Q2: Students who scored above 15

```python
above_15 = df_cleaned[df_cleaned['G3'] > 15]
count_above_15 = above_15.shape[0]
count_above_15
```

#### Q3: Correlation between study time and G3

```python
import numpy as np
```

```python
correlation = np.corrcoef(df_cleaned['studytime'], df_cleaned['G3'])[0,1]
correlation
```

#### Q4: Gender with higher average G3

```python
avg_by_gender = df_cleaned.groupby('sex')['G3'].mean()
avg_by_gender
```

### 5. Data Visualization

#### Histogram of G3:

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df_cleaned['G3'], bins=10, kde=True)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()
```

#### Scatter Plot: Study Time vs Final Grade

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='studytime', y='G3', data=df_cleaned)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

####  Bar Chart: Average G3 by Gender

```python
plt.figure(figsize=(6,4))
avg_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title("Average Final Grade (G3) by Gender")
plt.ylabel("Average G3")
plt.xticks(rotation=0)
plt.show()
```
